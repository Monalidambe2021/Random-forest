```python
# Import Library
import pandas as pd
```

```python
# Load the data
df= pd.read_csv('../../dnyan/Datasets/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
# Set independent & dependent variables
X=df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state= 46)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier ()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
```

```python
Params={'max_depth':[2,3,4,5],
'min_samples_leaf':[3,4,5,6],
'criterion':['gini','entropy'],
'n_estimators':[10,8,15,20,30,50,100,150]}
```

```python
grid_CV=GridSearchCV (RandomForestClassifier (),Params,cv=5, verbose=3)
grid_CV.fit(X,y)
```

```python
grid_CV.best_params_
```

```python
grid_CV.best_estimator_
```

```python
 model = RandomForestClassifier (max_depth=2,min_samples_leaf=3,n_estimators=10)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
